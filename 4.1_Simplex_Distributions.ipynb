# <font color="darkblue">Simplex Distibutions

The unit k-simplex is a set of points $x_i$ (real numbers in a k-tuple, $i = 1,2,..k$) such that

1. $x_i > 0$

1. $\sum\limits_{i=1}^kx_i = 1$

**$(0.35,0.22,0.23,0.12,0.08)$ is a 5-simplex**

**$(0.35,0.65)$ is a 2-simplex**

<font color="red">
1. Choose any five numbers between 1 and 100
1. Repeat 1, such that sum of the numbers chosen is 100

# <font color="darkblue"> Functional Form

## <font color="darkorange"> Bernoulli - Raw Observed Form - Dichotomous Catgorical

If $X \sim \mathrm{Bernoulli}(\theta)$ then, its PMF is

$$Pr(X=x)=\theta^x(1-\theta)^{1-x}$$ where $x=0,1$ and $0<\theta<1$

If we let, $\theta_1=Pr(X=C_1)$, then $\theta_2=Pr(X=C_2)$ such that $\theta_1+\theta_2=1$

## <font color="darkorange">Grouped - Binomial

<font color="maroon"> Genesis: Out of $N$ Bernoulli trials what is the number (proportion) of $C_i, i=1,2$

Hence, X is the **the number of $C_i$ out of N trials**

Then PMF of $X \sim \mathrm{Binomial}(N,\theta)$ is

$$Pr(X=x)= \binom{N}{x} \theta^x(1-\theta)^{N-x}$$ where $x=0,1\cdots\cdots N$ and $0<\theta<1$

Here again, if we let $\theta_1=Pr(X=C_1)$, then $\theta_2=Pr(X=C_2)$ such that $\theta_1+\theta_2=1$

## <font color="darkorange"> Multinoulli Raw Observed Form - Polychotomous Catgorical

If $X \sim \mathrm{Multinoulli}(\boldsymbol{\theta})$ then, its PMF is

$$Pr(\mathbf{X=x})=\theta_1^{x_1}\theta_2^{x_2}\cdots\cdots\theta_k^{x_k}$$ where $x_i=\{0,1\}$ and $0<\theta_i<1$ with $\sum\limits_{i=1}^k \theta_i = 1$

## <font color="darkorange">Grouped - Multinomial

<font color="maroon"> Genesis: Out of $N$ Multinoulli trials what is the number (proportion) of $C_i, i=1,2,3,\cdots\cdots,k$

Hence, X is the **the count (size) of each $C_i$ out of N trials, so that $\sum\limits_{i=1}^k x_i = N$**

If $X \sim \mathrm{Multinomial}(N,\boldsymbol{ \theta})$ then, its PMF is


$$\frac{N}{\prod\limits_{i=1}^k x_i!}\theta_1^{x_1}\theta_2^{x_2}\cdots\cdots\theta_k^{x_k}$$ where $x_i=\{0,1,2...k\}$ and $0<\theta_i<1$ with $\sum\limits_{i=1}^k \theta_i = 1$

#<font color="darkblue"> Sampling from Multinomial distribution

Assume a Multinoulli trial repeated n=20 times and looking for distribution counts of **$k$** classes in 20 trials, we simulate this from a multinomial distribution with a choice of $\boldsymbol{\theta}$

Let $k=5$

Let $\boldsymbol{\theta}=(0.35,0.15,0.05,0.40,0.10)$

We would like to simulate 10 items (no restrictions)

If $n=1$ this sampler will be values from a Multinoulli$\boldsymbol{(\theta)}$

```python
from scipy.stats import multinomial
n=1
theta=[0.35,0.15,0.05,0.40,0.10]
multinomial.rvs(n, theta, size=10, random_state=None)
```

# <font color="darkblue"> Bayesian Model

Assume a polychotomous variable with $k$ levels, and $n_1,n_2, \cdots n_k$ are the counts in each of the $k$ classes

Let $\theta_1,\theta_2, \cdots \theta_k$ be the proportion parameters for the class membership of the population

Two conditions are

1. $\sum n_j = N$, the total counts

2. $\sum \theta_j = 1$, total probability is one

In symbols,

$$Y \sim \mathrm{Multinomial}(\boldsymbol{\theta})$$ where $\boldsymbol{\theta}=(\theta_1, \theta_2, \cdots \theta_k)$

$$\boldsymbol{\theta} \sim \mathrm{Dirichlet}(\boldsymbol{\alpha})$$ where  $\boldsymbol{\alpha}=(\alpha_1, \alpha_2, \cdots \alpha_k)$

PDF of Dirichlet variate is $$\frac{\Gamma a}{\Gamma\alpha_1\Gamma\alpha_2\cdots\cdots\Gamma\alpha_k}\theta_1^{\alpha_1-1}\theta_2^{\alpha_2-1}\cdots\cdots\theta_k^{\alpha_k-1}$$

where $\sum \theta_j = 1$ and $a=\sum\alpha_j$

$\alpha'$s are called **concentration parameters**

[Wiki_Dirichlet](https://en.wikipedia.org/wiki/Dirichlet_distribution)
