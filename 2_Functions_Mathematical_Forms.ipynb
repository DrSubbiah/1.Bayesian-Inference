# <font color="darkblue"> Functions Basic Facts

There can be infinitely many ways to define mathematical form of a function;  however, such possibilities may be understood by realizing the relevance of defining functional form for describing numerous processes in the real world. Surprisingly this is possible with five basic functional forms and five operations $+,~-,~\times,~\div,$and $\circ$; last one $\circ$ indicates <font color="red"> **composition of functions**



1. **Constant Function** $f(x)=k$

1. **Power Function** $f(x) = x^k$, except at $x=0$

  - If $k = 1$ then $f(x)$ is linear, a straight line

1. **Exponential Function** $f(x) = k^x$ with $k>0$

  - In particular, $k=e$ draws much scientific attention

1. **Logarithmic Function** $f(x) = \log_{a}x$ where $a>0$ is the base  

   - Common logarithm: $a=10$
   
   - Natural logarithm: $a=e$

1. **Circular Trigonometric Function**

  - Sine

  - Cosine

```python
import numpy as np
import pandas as pd
import scipy as sc
from scipy import stats as st
import math
import matplotlib.pyplot as plt
```

# <font color="darkblue"> Power Functions

```python
# Creating vectors X

x = np.linspace(-2, 2, 100)

def f1(x):
    return x ** 2
f12 = np.vectorize(f1)

def f2(x):
    return 1 - x**2
f22 = np.vectorize(f2)

def f3(x):
    return (x-1)**2
f32 = np.vectorize(f3)

def f4(x):
    return (x**2)/2
f42 = np.vectorize(f4)

def f5(x):
    return 2*(x**2)
f52 = np.vectorize(f5)

fig = plt.figure(figsize = (20, 8))

plt.plot(x, f12(x),label="$x^2$",color="magenta")
# plt.plot(x, f22(x),label="$1-{x}^2$")
# plt.plot(x, f32(x),label="${(x-1)}^2$")
plt.plot(x, f42(x),label="$(x^2)/2$")
plt.plot(x, f52(x),label="$2x^2$")


plt.title('Power Functions $x^k$',fontsize=20)
plt.legend(fontsize=15)
plt.axhline(y=0, color='g', linestyle='--')

plt.axvline(x=0, color='r', linestyle=':')
plt.axvline(x=1, color='r', linestyle=':')

plt.show()
```

# <font color="darkblue"> Exponential Functions

```python
x = np.linspace(-2, 2, 100)

def f1(x):
    return math.exp(x)
f12 = np.vectorize(f1)

def f2(x):
    return 1.5**x
f22 = np.vectorize(f2)

def f3(x):
    return 5**x
f32 = np.vectorize(f3)

fig = plt.figure(figsize = (20, 8))

plt.plot(x, f12(x),label="$e^{x}$")
plt.plot(x, f22(x),label="$1.5^{x}$")
plt.plot(x, f32(x),label="$5^{x}$")

plt.axvline(x=0, color='r', linestyle=':')
plt.title('Exponential Functions $a^x$',fontsize=20)
plt.legend(fontsize=20)
plt.show()
```

# <font color="darkblue"> logarithmic Functions

```python
x = np.linspace(0.000001, 10, 100)

def f1(x):
    return math.log(x)
f12 = np.vectorize(f1)

def f2(x):
    return math.log(8*x)
f22 = np.vectorize(f2)

def f3(x):
    return math.log(x)+8
f32 = np.vectorize(f3)

def f4(x):
    return math.log(x+1)
f42 = np.vectorize(f4)

fig = plt.figure(figsize = (20, 8))

plt.plot(x, f12(x),label="log(x)")
# plt.plot(x, f22(x),label="log(8x)")
# plt.plot(x, f32(x),label="log(x)+8")
# plt.plot(x, f42(x),label="log(x+1)")


plt.title('Logarithmic Functions $log_e(x)$ - base e',fontsize=20)
plt.legend(fontsize=20)
plt.show()
```

# <font color="darkblue"> Compostion of Functions

```python
x = np.linspace(2, 5, 100)
def f1(x):
    return math.log(x**2-1)
f12 = np.vectorize(f1)

def f2(x):
    return math.log(2*x**2)
f22 = np.vectorize(f2)


def f3(x):
    return math.log((x-1)**2)
f32 = np.vectorize(f3)

fig = plt.figure(figsize = (20, 8))

plt.plot(x, f12(x),label="$log(x^2-1)$")
plt.plot(x, f22(x),label="$log(2x^2)$")
plt.plot(x, f32(x),label="$log{(x-1)}^2$")



plt.title('Composition of Functions',fontsize=20)
plt.legend(fontsize=20)
plt.show()
```

# <font color="darkblue"> Derived from Logistic Function

$$\frac{e^{ax}}{1+e^{ax}}$$ where $-\infty < x <\infty$

This has a nice property of having restricted within $(0,1)$



```python
x = np.linspace(-10, 10, 100)

def f1(x):
    return math.exp(x)/(1+math.exp(x))
f12 = np.vectorize(f1)

def f2(x):
    return math.exp(x/2)/(1+math.exp(x/2))
f22 = np.vectorize(f2)

def f3(x):
    return math.exp(2*x)/(1+math.exp(2*x))
f32 = np.vectorize(f3)

def f4(x):
    return math.exp(2*(x-1))/(1+math.exp(2*(x-1)))
f42 = np.vectorize(f4)

fig = plt.figure(figsize = (20, 8))
plt.plot(x, f12(x),label="logi(x)")
# plt.plot(x, f22(x),label="logi(x/2)")
# plt.plot(x, f32(x),label="logi(2x)")
plt.plot(x, f42(x),label="logi(2(x-1))")

plt.title('Composition - Logistic Function',fontsize=20)
plt.legend(fontsize=20)
plt.show()
```
